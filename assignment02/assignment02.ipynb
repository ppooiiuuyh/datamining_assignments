# import packages for plotting graphs and manipulating data:

```python
import numpy as np
import matplotlib.pyplot as plt
```

# define my function: $f(x) = \cos(x) \cdot x$

```python
def myFunction(x):
    f = np.cos(x) * x
    return f
```

# define the derivative of my function: $first : f'(x) = -\sin(x) \cdot x + \cos(x)$ $second : f''(x) = -\cos(x) \cdot x - 2\sin(x)$

```python
def myDerivativeFunction(x):
    Df = - np.sin(x) * x + np.cos(x)
    return Df

def myDerivativeFunction_2(x):
    Df2 = -np.cos(x)*x -np.sin(x) + -np.sin(x)
    return Df2
```

# define second order Taylor expansion

```python
def seconde_order_taylor(x,a,func, first_der, second_der):
    sot = func(a) + (x-a)*first_der(a) + ((x-a)**2)*second_der(a)/2
    return sot
```

# define the domain of the function: $x = [-15 : 0.1 : 15]$

```python
x = np.arange(-15, 15, 0.1)
A = [0,10,-5]
```

# compute the graph

```python
f = myFunction(x)
Df = myDerivativeFunction(x)

sots = []
for a in A:
    sots.append(seconde_order_taylor(x,a,myFunction,myDerivativeFunction,myDerivativeFunction_2))
```

# plot the graphs for the function and its derivative

```python
plt.figure(1)
plt.ylim(-40,40)
plt.plot(x, f, 'b', label="function")
#plt.plot(x, Df, 'r', label="derivative")
for e,a in enumerate(A) :
    plt.plot(x, sots[e], label="taylor_"+str(a))
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.show()
```
